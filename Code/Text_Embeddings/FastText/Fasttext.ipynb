```python
import fasttext
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv("Datasets/amazon_reviews_3.csv")
```

```python
texts = df['PREPROCESSED_REVIEW_TEXT']
labels = df['LABEL_ENCODED']
```

```python
X_train, X_test, y_train, y_test = train_test_split(texts, labels, 
                                                   test_size=0.2,
                                                   random_state=42)
```

```python
#CREATE A FILE WITH FORMAT: __label__ positive This movie is great!
```

```python
with open("Datasets/fasttext_training.txt", "w") as file:
    for text, label in zip(X_train, y_test):
        file.write(f'__label__{label} {text}\n')
```

```python
#TRAIN THE MODEL
model = fasttext.train_supervised('Datasets/fasttext_training.txt')
```

```python
#USE THE MODEL FOR PREDICTIONS
text_sample = "amazing wow love it"
predicted_label = model.predict(text_sample)
```

```python
predicted_label
```

```python
predictions = []
for text in X_test:
    prediction = model.predict(text)[0][0]
    predictions.append(int(prediction[-1]))
    
predictions
```

```python
y_test
```

```python
print(accuracy_score(y_test, predictions))
```
