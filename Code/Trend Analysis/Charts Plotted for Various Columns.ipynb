```python
#HEATMAP FOR SENTIMENT SCORE
pivot_table = df.pivot_table(values = 'REVIEW_SENTIMENT', index = 'Unnamed: 0.4')
```

```python
sns.heatmap(pivot_table, cmap='coolwarm', annot=True, fmt=".2f")
```

```python
#BAR CHART
sentiment_counts = df['REVIEW_SENTIMENT'].value_counts()
```

```python
sentiments = sentiment_counts.index
counts = sentiment_counts.values
```

```python
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.title('Review Sentiment Bar Chart')
plt.bar(sentiments, counts)
```

```python
#RATING DEVIATION
rating_deviations = df['RATING_DEVIATION']
plt.xlabel('Rating Deviation')
plt.ylabel('Frequency')
plt.title('Rating Deviation Distribution')

plt.hist(rating_deviations, bins=10) 
```

```python
#MAX_REVIEWS_DAY
rating_deviations = df['MAX_REVIEWS_DAY']
plt.xlabel('Maximum Reviews in a day')
plt.ylabel('Frequency')
plt.title('Maximum Reviews Distribution')

plt.hist(rating_deviations, bins=10) 
```

```python
#TOTAL USER REVIEWS
user_reviews = df['TOTAL_USER_REVIEWS']
plt.xlabel('Total user reviews')
plt.ylabel('Frequency')
plt.title('Total User Reviews Distribution')

plt.hist(rating_deviations, bins=10) 
```

```python
#TOTAL USER HELPFUL VOTES
helpful_votes = df['TOTAL_USER_REVIEWS']
plt.xlabel('Total user helpful votes reviews')
plt.ylabel('Frequency')
plt.title('Total Helpful Votes Distribution')

plt.hist(rating_deviations, bins=10) 
```
