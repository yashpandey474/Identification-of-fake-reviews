```python
#IMPORT NECESSARY LIBRARIES
import pandas as pd
import numpy as np
from nltk.stem import PorterStemmer
import spacy

from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import IsolationForest
```

```python
df = pd.read_csv("Datasets/amazon_reviews_3.csv")
```

```python
df.columns
```

```python
X = df[[
    'RATING',
       'VERIFIED_PURCHASE', 'NUM_NOUNS',
       'NUM_VERBS', 'NUM_ADJECTIVES', 'NUM_ADVERBS', 'REVIEW_LENGTH',
        'TITLE_LENGTH'
    
]]
```

```python
#GRID SEARCH FOR OPTIMAL N_ESTIMATORS
model = IsolationForest(random_state = 42)

#POSSIBLE VALUES
param_grid = {'n_estimators': [25, 30, 40, 50, 70, 100,  150, 175, 200]}

#GRID CREATION
grid_search = GridSearchCV(model, param_grid, cv = 5, scoring='accuracy')

#FIT THE GRID ON FEATURES
grid_search.fit(X)

best_n_estimators = grid_search.best_params_['n_estimators']
```

```python
best_n_estimators
```

```python
model = IsolationForest(n_estimators = 100, random_state = 42)

model.fit(X)

anomaly_scores = model.decision_function(X)

labels = model.predict(X)
```
